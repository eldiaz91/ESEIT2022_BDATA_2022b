# Visualizacion de datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Diagramas de dispersion (scatterplot)
# un scatterplot necesita una variable x y una variable y para graficar

x=np.arange(2,20)
y=x**2

plt.scatter(x,y, color='Purple' , marker='+')
plt.title('Funcion de x f(x)= x^2', font='courier new', fontsize=20, color = 'green') #Titulo .title, tamaño de fuente, color de fuente
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()
```

```python
#Agregando mas scatter
x=np.arange(2,20)
y=x**3
z=x**0.5

plt.scatter(x,y, color='Purple' , marker='+', label ='x^2')
plt.scatter(x,z, color='Red' , marker='s', label = 'x^3')
plt.title('Funciones Polinómicas', font='courier new', fontsize=20, color = 'green') #Titulo .title, tamaño de fuente, color de fuente
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()


plt.text(1.5,5500, 'texto', color = 'blue')
```

```python
# Histogramas
mean = 5
std = 3
n= 10

x=np.random.normal(mean, std, n) #ultimo valor es la cantidad de numeros a generar
#print(x)

plt.hist(x, bins=20);  #la ; es para que no imprima los numeros que estan en el arreglo

plt.title('Histograma de una variable aleatoria', font='courier new', fontsize=20, color = 'green') #Titulo .title, tamaño de fuente, color de fuente
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.text(4,1600, f'Mean:'{mean}, color = 'blue')
```

```python
plt.boxplot(x, vert=False)
```

```python
#Subplots, defino una figura y los ejes (derecha, arriba, abajo, etc) es como un lienzo

fig, ax = plt.subplots(ncols=2, nrows=1, figsize=(15,5)) #2 columnas, 1 fila, pixeles 15 y 5
#ax es un arreglo, una lista con los subplots que se estan creando

ax[0].hist(x) #a la izquierda el histograma de x
ax[1].boxplot(x, vert=False);
```

```python
#Cambiando la orientacion

fig, ax = plt.subplots(ncols=1, nrows=2, figsize=(5,10), sharex=True) #2 Shareex= compartan el eje x, los datos de ex solo aparecen una vez abajo
#ax es un arreglo, una lista con los subplots que se estan creando

ax[0].hist(x) #a la izquierda el histograma de x
ax[1].boxplot(x, vert=False);
```
